                                                         Exploratory Data Analysis

                                                              GENDER


```python
df.groupby('Gender').size().plot(kind='pie', autopct='%.2f')
```

From Above Graph we can see that most of the buyers are Females

```python
Sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum()
barplot = sns.barplot(x = 'Gender',y = 'Amount',data = Sales_gen)
#for bar in barplot.containers:
 # barplot.bar_label(bar)
```

From above graph it is clear that Purchasing power of Female is greater than Men

                                                                AGE

```python
x = sns.countplot(data = df, x = 'Age Group',hue = 'Gender')
for bars in x.containers:
    x.bar_label(bars)
```

```python
Sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

plt.pie(Sales_age['Amount'], labels=Sales_age['Age Group'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sales Distribution by Age Group')
plt.show()
```

By observing the graphs from above, it becomes evident that the majority of buyers fall within the age range of 26 to 35 years and are female.

                                                                STATE

```python
Sales_State = df.groupby(['State'])['Orders'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(10, 8))
Sales_State.sort_values().plot(kind='barh')
plt.xlabel('Total Orders')
plt.ylabel('State')
plt.title('Top 15 States by Total Orders')
plt.tight_layout()
plt.show()
```

```python
Sales_State = df.groupby(['State'])['Amount'].sum().sort_values(ascending=False).head(15)

plt.figure(figsize=(8, 8))
plt.pie(Sales_State, labels=Sales_State.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Total Sales Amount Among Top 15 States')
plt.show()
```

The graphs above highlight Uttar Pradesh, Maharashtra, and Karnataka as the leading contributors, both in terms of order count and total sales amount.

```python
a = sns.countplot(data = df , x = 'Marital_Status')
sns.set(rc = {'figure.figsize':(6,5)})
for bars in a.containers:
       a.bar_label(bars)
```

```python
plt.figure(figsize=(10, 6))

Sales_State = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().unstack()

# Rename marital status values for better labeling
Sales_State.rename(index={1: 'Single', 0: 'Married'}, inplace=True)

a = Sales_State.plot(kind='bar', stacked=True)
a.set_xlabel('Marital Status')
a.set_ylabel('Total Sales Amount')
a.set_title('Total Sales Amount by Marital Status and Gender')
plt.legend(title='Gender', labels=['Female', 'Male'], loc='upper right')
plt.tight_layout()

plt.show()
```

The graphs present a noticeable pattern: a considerable number of buyers consist of married women,and they have high purchasing power.

                                                occupation

```python

# Set figure size
plt.figure(figsize=(20, 5))

# Create a countplot of occupations
ax = sns.countplot(data=df, x='Occupation')

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Customize plot appearance
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Count of Occupations')

# Show the plot
plt.show()
```

```python
# Set figure size
plt.figure(figsize=(15, 5))
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
# Create a swarm plot of total sales amount for different occupations
sns.swarmplot(data=sales_state, y='Amount', x='Occupation')

# Customize plot appearance
plt.xlabel('Occupation')
plt.ylabel('Total Sales Amount')
plt.title('Total Sales Amount by Occupation (Swarm Plot)')

# Rotate occupation labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

The graphs clearly show that a large number of buyers work in the IT, Healthcare, and Aviation sectors.

                                                       Product Category

```python
# Set figure size
plt.figure(figsize=(24, 10))

# Create a bar plot of product categories with bar labels
ax = sns.countplot(data=df, x='Product_Category')

# Add labels to the bars
for bars in ax.containers:
    ax.bar_label(bars)

# Customize plot appearance
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.title('Count of Products in Each Category')

# Show the plot
plt.show()
```

```python

# Calculate the sum of amounts for top 10 product categories
top_sales_categories = df.groupby(['Product_Category'])['Amount'].sum().sort_values(ascending=False).head(10)

# Set figure size
plt.figure(figsize=(8, 8))

# Create a pie chart for top 10 product categories by total sales amount
plt.pie(top_sales_categories, labels=top_sales_categories.index, autopct='%1.1f%%', startangle=140)

# Customize plot appearance
plt.title('Distribution of Total Sales Amount by Top 10 Product Categories')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```

The above visual representations clearly indicate that a majority of the sold products originate from the Food, Clothing, and Electronics categories.

```python
# Calculate the top 10 most sold products
top_sold_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10)

# Create a bar plot for top 10 most sold products
fig, ax = plt.subplots(figsize=(10, 6))
top_sold_products.sort_values().plot(kind='barh', color='skyblue')

# Add labels, title, and customize appearance
plt.xlabel('Total Orders')
plt.ylabel('Product ID')
plt.title('Top 10 Most Sold Products')

# Show the plot
plt.show()
```

                                                          **Conclusion**

Married women within the age group of 26-35 years, hailing from the states of Uttar Pradesh, Maharashtra, and Karnataka, and engaged in the IT, Healthcare, and Aviation sectors, exhibit a higher propensity for purchasing products falling under the Food, Clothing, and Electronics categories.
